## Tarea 1

```python
strategies=['Rock','Paper','Scissors']
```

```python
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

```python
names=['Jim','Jane','Peter','Zoe']
```

```python
society=[{'name':n,'score':0,'strategy':None} for n in names]
```

```python
import itertools

# pair is a tuple of dicts
for pair in itertools.combinations(society,2):
    print(pair)
```

```python
import itertools

# each dict
for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

```python
from random import choice
choice(strategies);
```

```python
society=[{'name':n,'score':0,'strategy':None} for n in names]

# several rounds
for aRound in range(100): # juegan 100 veces

    # en each round:
    for player1,player2 in itertools.combinations(society,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        # result from strategy chosen
        result=payoff[player1['strategy'],player2['strategy']]


        # update scores
        player1['score']+=result[0]
        player2['score']+=result[1]

        # print()
        print(player1['name'], player2['name'])
        print(player1['strategy'], player2['strategy'])
        print(player1['score'], player2['score'])
        print('---------------------')
```

```python
society
```

```python
import pandas as pd

socialResults=pd.DataFrame(society)
socialResults
```

```python
winnerScore=socialResults.score.max()

socialResults[socialResults.score==winnerScore]
```

## Tarea 3: Abstraction of agents
### "Virus"

```python
from random import choice
```

```python
strategies=['Virus', 'Medicina', 'Vacuna']
```

```python
payoff={('Virus','Medicina'):(1,0), # Infecta
        ('Medicina','Virus'):(0,1), 
        ('Medicina','Vacuna'):(2,0), # cura y fortalece
        ('Vacuna','Medicina'):(0,2),
        ('Vacuna','Virus'):(2,0), # protege
        ('Virus','Vacuna'):(0,2),
        ('Virus','Virus'):(0,0), # neutralizan
        ('Medicina','Medicina'):(1,1), # ambos curan
        ('Vacuna','Vacuna'):(0,0)} # protección sin daño
```

```python
class Jugador:
    def __init__(self, name, score=0): 
        self.name = name                
        self.score = score               
        self.strategy = None            

    def increase_score(self, value):  
        self.score += value

    def get_strategy(self):
        strategies = ['Virus', 'Medicina', 'Vacuna']
        self.strategy = choice(strategies)  
        return self.strategy
```

```python
jugadoras = [
    Jugador("Brisa"),
    Jugador("Jimena"),
    Jugador("Micaela"),
    Jugador("Fabiana"),
    Jugador("Kaylee")]
```

```python
def play_round(jugador1, jugador2):
    strategy1 = jugador1.get_strategy()  
    strategy2 = jugador2.get_strategy()  
    
    payoff1, payoff2 = payoff[(strategy1, strategy2)]  
    
    jugador1.increase_score(payoff1)  
    jugador2.increase_score(payoff2)  

    # nombre, estrategia y puntaje de las jugadoras
    print(jugador1.name, jugador2.name)
    print(jugador1.strategy, jugador2.strategy)
    print(jugador1.score, jugador2.score)
    print()
```

```python
for i in range(100):
    print('round:',i+1)
    
    for j in range(len(jugadoras)):
        for k in range(j+1, len(jugadoras)):
            play_round(jugadoras[j], jugadoras[k])  
    print("-----------------------")
```

```python
print("Puntaje total")
for jugadora in jugadoras:
    print(f"{jugadora.name} {jugadora.score}")
```
